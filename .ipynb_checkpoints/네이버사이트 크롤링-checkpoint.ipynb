```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('#productListArea > ul > li > p > a')
for item in items:
    print(item.get_text())
```

* '#productListArea > ul > li:nth-child(1) > p > a'
* '#productListArea > ul > li:nth-child(2) > p > a'
* -> '#productListArea > ul > li:nth-child > p > a' 가 반복됨을 유추

* https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000 ; 패션의류
* https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001 ; 패션잡화
* https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002 ; 화장품미용
* 사이트 주소도 뒤에 숫자만 조금씩 바뀜

```python
import requests
from bs4 import BeautifulSoup
site_list = ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003']
# 사이트를 리스트로 만들기
for site in site_list:
# 리스트를 for문으로 반복    
    res = requests.get(site)
    soup = BeautifulSoup(res.content, 'html.parser')
    items = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for item in items:
        print(item.get_text())
```

```python
import requests
from bs4 import BeautifulSoup
site_list= ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003']
for site in site_list:
    res = requests.get(site)
    soup = BeautifulSoup(res.content,'html.parser')
    items = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for index, item in enumerate(items) :
## 반복 출력되는 아이템에 인덱스 출력하기
        print(str(index +1) + '.', item.get_text())
```

```python
## 딕셔너리로 주소 인덱스 출력하기(숙제)

import requests
from bs4 import BeautifulSoup
site_dict= {1:'https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000',2:'https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001',3:'https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002',4 :'https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003'}
for site in site_dict:
    res = requests.get(site)
    soup = BeautifulSoup(res.content,'html.parser')
    items = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for index, item in enumerate(items) :
        print(str(index +1) + '.', item.get_text())
```

#### 해외 종목 top 10
* 웹페이지 : https://finance.naver.com/sise/
* 추출할 텍스트 :
* #contentarea > div.box_top_submain2 > div.rgt > ul.lst_major > li:nth-child(1) > a  ; 다우산업
* #contentarea > div.box_top_submain2 > div.rgt > ul.lst_major > li:nth-child(2) > a  ; HMM
* #contentarea > div.box_top_submain2 > div.rgt > ul.lst_major > li > a 가 반복

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/sise')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('#contentarea > div.box_top_submain2 > div.rgt > ul.lst_major > li > a')
for item in items:
    print(item.get_text())
```

```python
## 지수이름과 지수, 상승/하락 표시하기
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/sise')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('div.rgt > ul.lst_major > li')
for item in items:
    print(item.get_text())
```

```python
## 텍스트도 함께 표시하기
## 지수명 추출 ; div.rgt > ul.lst_major > li > a
## 지수 추출 ; div.rgt > ul.lst_major > li > span
## 상승/하락 추출 ; div.rgt > ul.lst_major > li > em > span
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/sise')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('div.rgt > ul.lst_major > li')
for item in items:
    print("지수이름 : ", item.find('a').get_text(), '\n', "현재지수 : ", item.find('span').get_text(),item.find('em').get_text()) 
```
