```python
import requests
from bs4 import BeautifulSoup

service_key = 'jb%2F7BI2zttw%2B6nUHthsazwWGPlZUS49XI0zgb%2Bdx68OF7M1vRJK%2B7tBwZ2AHUtQLzIU%2BymHLt1LyqjVp1Q5gwA%3D%3D'
params = '&numOfRows=10&pageNo=1&sidoName=부산'
open_api = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty?serviceKey='
api = open_api + service_key + params
res = requests.get(api)
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.find_all('item')

for item in data:
#     print(item)
    stationname = item.find('stationname')
    pm10grade = item.find('pm10grade')
    print(stationname.get_text(), pm10grade.get_text())
```
